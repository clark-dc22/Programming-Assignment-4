```python
import pandas as pd
```

```python
df = pd.read_csv('board2.csv')
df
```

```python
df.describe()
```

```python
#Code that displays only students from Visayas and has a score less than 70.
Vis = df.loc[(df['Hometown'] == 'Visayas')&(df['Math']<70),['Name','Gender','Track','Math']] 
Vis
```

```python
#Code that displays students from Luzon, has a track of Instrumentation, and has a score greater than 70 in the track, Electronics.
Instru = df.loc[(df['Hometown'] == 'Luzon') & (df['Track'] == 'Instrumentation') & (df['Electronics'] > 70), ['Name','GEAS','Electronics']]
Instru
```

```python
#Code that displays female students from Mindanao and has an average greater than or equal to 55.
df['Average'] = df[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1) 

df.loc [(df['Gender'] == 'Female') & (df['Hometown'] == 'Mindanao') & (df['Average'] >= 55), ['Name', 'Track', 'Electronics', 'Average']] 
```

```python
import matplotlib.pyplot as plt
```

```python
#Code that displays the average by each track.
df['Average'] = df[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1) 

avg = df.groupby('Track')['Average'].mean() 

plt.bar (avg.index, avg.values, color = ['red', 'blue', 'yellow']) 
plt.title('Average by Track') 
plt.ylabel('Mean') 
plt.xlabel('Tracks') 
plt.show() 
```

```python
#Code that displays the average of each gender from each hometown.
plt.bar (avg.index, avg.values, color = ['red', 'blue', 'yellow']) 
plt.title('Average by Gender') 
plt.ylabel('Mean') 
plt.xlabel('Gender') 
plt.show() # display
```

```python
#Code that shows average of each hometown.
avg = df.groupby('Hometown')['Average'].mean() 

plt.bar (avg.index, avg.values, color = ['red', 'blue', 'yellow']) 
plt.title('Average by Hometown') 
plt.ylabel('Mean') 
plt.xlabel('Hometown') 
plt.show() 
```
